```python
# 1. Write a Python program to find the sum of all elements in the list [10, 20, 30,40, 50].

def sum(lis):



    elm_sum=0
    for i in range(len(lis)):
        elm_sum+=lis[i]
    return elm_sum

    
lis=[10, 20, 30,40, 50]
sum(lis)
```

```python
# 2. Write a Python program to display the odd and even elements from the list [10,23, 11, 12, 33, 44, 2, 5, 6].

lis = [10,23, 11, 12, 33, 44, 2, 5, 6]

odd=[]
even=[]
for i in range(len(lis)):
    if i%2==0:
        even.append(i)
    else:
        odd.append(i)
print(f"list for odd num is  : {odd}")
print(f"list for even num is : {even}")
```

```python
# 16.Write a Python program to print all the vowels present in the given list of strings ["Dreamer", "infotech"].

string =["Dreamer", "infotech"]
new_str=""

for i in string:
    for j in i :
        if j.lower() in "aeiou":
            new_str+= j
print(f" the all volwels from the string {string} are : {new_str}")
    
```

```python
# 28. Create a new list containing only prime numbers from a given list. 

lis = [1, 2, 3, 4, 56, 1, 22, 23, 33, 23, 56]
new_lis=[]

for i in lis:
    if i<2:
        continue
    count=0
    for j in range(1,i+1):
        if i%j==0:
            count+=1
    if count==2:
        new_lis.append(i)

print(new_lis)
```
