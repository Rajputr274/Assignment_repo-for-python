```python
# 4. Write a Python program to interchange the first and last elements in the list [10,23, 11, 12, 33, 44, 2, 5, 6].

def interchange(lis):
    lis = [10,23, 11, 12, 33, 44, 2, 5, 6]
    firs_elm=lis[-1]
    last_elm=lis[0]
    lis[0]=firs_elm
    lis[-1]=last_elm

    return lis
interchange(lis)
```

```python
# 22. Flatten a nested list like [[1,2], [3,4], [5,6]] into a single list. 
def flatten():

    lis=[[1,2], [3,4], [5,6]]
    l2=[]
    for i in lis:
        l2.extend(i)
    return l2

flatten()
```

```python
# 28. Create a new list containing only prime numbers from a given list. 
def lis_prime():
    lis = [1, 2, 3, 4, 56, 1, 22, 23, 33, 23, 56]
    new_lis=[]

    for i in lis:
        if i<2:
            continue
        count=0
        for j in range(1,i+1):
            if i%j==0:
                count+=1
        if count==2:
            new_lis.append(i)

    return new_lis
lis_prime()
```

```python
# 36. Rearrange a list so that positive and negative numbers alternate.
# 
def rearrange():

    lis=[1, 2, -3, -4, 5, -6]
    positive=[]
    negative=[] 
    for i in lis:
        if i>0:
            positive.append(i)
        else:
            negative.append(i)
    new_lis=[]
    for i in range(max(len(positive),len(negative))):
        if i<len(positive):
            new_lis.append(positive[i])
        if i<len(negative):
            new_lis.append(negative[i])
    return new_lis
rearrange()
```

```python
# 37. Find all pairs of elements in a list whose sum equals a given target number. 
def pairs():

    List = [2, 4, 3, 5, 7, 8, 9] 
    Target = 7 
    pairs = []
    for i in range(len(List)):
        for j in range(i+1, len(List)):
            if List[i] + List[j] == Target:
                pairs.append((List[i], List[j]))
    return pairs
pairs()
```

```python
# 42. Find All Anagram Groups in a List of Words. 

def anagram():

    List=["eat", "tea", "tan", "ate", "nat", "bat"]
    anagram_groups = {} 
    for word in List:
        sorted_word = ''.join(sorted(word))
        if sorted_word in anagram_groups:
            anagram_groups[sorted_word].append(word)
        else:
            anagram_groups[sorted_word] = [word]
    return anagram_groups
anagram()
```
